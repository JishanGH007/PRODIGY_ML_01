```python
import pandas as pd
```

```python
train_data = pd.read_csv('C:/Users/JISHAN/Downloads/prodigy/house-prices-advanced-regression-techniques/train.csv')
test_data=pd.read_csv('C:/Users/JISHAN/Downloads/prodigy/house-prices-advanced-regression-techniques/test.csv')
test_data_price=pd.read_csv('C:/Users/JISHAN/Downloads/prodigy/house-prices-advanced-regression-techniques/sample_submission.csv')
```

```python
test_data_price.head()
test_data_price['SalePrice'] = test_data_price['SalePrice'].astype(int)
test_data_price.head()
```

```python
train_data.head()
test_data.head()
```

```python
print(train_data.isnull().sum())
print(test_data.isnull().sum())
```

```python

numeric_cols1 = train_data.select_dtypes(include=['number']).columns
numeric_cols2 = test_data.select_dtypes(include=['number']).columns
train_data[numeric_cols1] = train_data[numeric_cols1].fillna(train_data[numeric_cols1].mean())
test_data[numeric_cols2] = test_data[numeric_cols2].fillna(test_data[numeric_cols2].mean())

print(train_data.dtypes)
print(train_data.dtypes)
print(test_data.dtypes)
```

```python
train_data['bathrooms']=train_data['BsmtFullBath']+train_data['BsmtHalfBath']+train_data['FullBath']+train_data['HalfBath']
test_data['bathrooms']=test_data['BsmtFullBath']+test_data['BsmtHalfBath']+test_data['FullBath']+test_data['HalfBath']
train_data['sq_foot']=train_data['GrLivArea']+train_data['TotalBsmtSF']
test_data['sq_foot']=test_data['GrLivArea']+test_data['TotalBsmtSF']
```

```python
train_data.head()
```

```python
test_data.head()
```

```python
test_data['sq_foot'] = test_data['sq_foot'].astype(int)
test_data['bathrooms'] = test_data['bathrooms'].astype(int)
```

```python
test_data.head()
```

```python
test_data['SalePrice'] = test_data_price.iloc[:, 1]
```

```python
test_data.head()
```

```python
columns_to_retain = ['sq_foot', 'BedroomAbvGr', 'bathrooms', 'SalePrice']

train_reduced = train_data[columns_to_retain]
test_reduced= test_data[columns_to_retain]
```

```python
train_reduced
```

```python
test_reduced
```

```python
from sklearn.preprocessing import PowerTransformer
```

```python
pt = PowerTransformer(method='yeo-johnson', standardize=True)
features_to_transform = ['sq_foot','BedroomAbvGr','bathrooms','SalePrice']
transformed_data_train = pt.fit_transform(train_reduced[features_to_transform])
transformed_data_test = pt.fit_transform(test_reduced[features_to_transform])
```

```python
trans_train_df = pd.DataFrame(transformed_data_train, columns=features_to_transform)
trans_test_df = pd.DataFrame(transformed_data_test, columns=features_to_transform)

df_transformed_train = pd.concat([trans_train_df, train_reduced.drop(columns=features_to_transform)], axis=1)
df_transformed_test = pd.concat([trans_test_df, test_reduced.drop(columns=features_to_transform)], axis=1)
print(df_transformed_train)
print(df_transformed_test)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
```

```python
x_train=df_transformed_train[['sq_foot','BedroomAbvGr','bathrooms']]
x_test=df_transformed_test[['sq_foot','BedroomAbvGr','bathrooms']]
```

```python
y_train=df_transformed_train[['SalePrice']]
y_test=df_transformed_test[['SalePrice']]
```

```python
x_train.shape
```

```python
y_train.shape
```

```python
x_train
```

```python
y_train
```

```python
model = LinearRegression()
model.fit(x_train, y_train)
y_pred = model.predict(x_test)

```

```python
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)
print("Mean Squared Error (MSE):", mse)
print("R-squared (R2):", r2)
```
